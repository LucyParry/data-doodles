```python
# Standard imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('ggplot')
plt.rc('figure')
```

```python
cat_data = { 
    'Cat': ['@Number10Cat', '@Diplomog', '@HMCabinetcat', '@TreasuryMog', '@CabOfficeMogs'], 
    'Conservative': [33, 38, 18, 10, 6], 
    'Labour': [33, 9, 4, 7, 2], 
    'Other': [11, 5, 1, 2, 0] 
}

cat_df = pd.DataFrame(cat_data, columns = ['Cat', 'Conservative', 'Labour', 'Other'])
cat_df
```

```python
pos = list(range(len(cat_df['Conservative']))) 
width = 0.25 
    
fig, ax = plt.subplots(figsize=(12, 6))

plt.bar(pos, cat_df['Conservative'], width, color='#0000FF', alpha = 0.8, label = cat_df['Cat'][0]) 
plt.bar([p + width for p in pos], cat_df['Labour'], width, color='#cc0000', alpha = 0.8, label = cat_df['Cat'][1]) 
plt.bar([p + width * 2 for p in pos], cat_df['Other'], width, color='#ff9900', alpha = 0.8, label = cat_df['Cat'][2]) 

plt.title('Government Cats of Twitter - Followers by Party', size = 20, fontname = "Arial", fontweight = "bold")

ax.set_ylabel('Followers', size = 13, fontname = "Arial", fontweight = "bold")
ax.set_xticks([p + 1.5 * width for p in pos])

ax.set_xticklabels(cat_df['Cat'], size = 13, fontname = "Arial", fontweight = "bold")
ax.grid('off', which='major', axis='y')

plt.xlim(min(pos) - width, max(pos) + width * 4)
plt.ylim([0, 55])

# Legend
plt.legend(['Conservative', 'Labour', 'Other'], loc='upper left', fontsize=12)

# Data source footer
source_string = 'Data from the (excellent) tool by @Psythor published by Politico\nAvailable at: https://www.politico.eu/article/westminster-twitter-bubble-illustrated-guide-graphics/'
plt.gcf().text(0.13, -0.01, source_string, fontsize=10)

plt.grid()
plt.show()
plt.savefig('politicats.png')
```
