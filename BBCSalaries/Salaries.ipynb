```python
# Standard data analysis imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import gridspec
import matplotlib.patches as mpatches

# Setting plots to use ggplot and some basic defaults for titles etc.
%matplotlib inline
plt.style.use('ggplot')
plt.rc('font', size=20) 
plt.rc('axes', titlesize=15)
plt.rc('axes', labelsize=10)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=15)
plt.rc('figure', titlesize=15)
```

```python
# Detecting the character encoding (I got an error message on the first tried with the default UTF-8) 
!/usr/local/bin/chardetect 'data/bbc_salaries_on_air.csv'
```

```python
# Read in the data (using ISO-8859-1 in the end because ISO-8859-2 gave me the wrong sign for the £)
salary_df = pd.read_csv('data/bbc_salaries_on_air.csv', encoding = "ISO-8859-1")
salary_df.head(5)
```

```python
# Sorting by salary and name
salary_df = salary_df.sort(['Band Upper Bound', 'Name'], ascending=False)
salary_df
```

```python
# A quick look at the descriptive statistics (and a quick way to find the total records) 
salary_df.describe()
```

```python
# Calculating the percentage of Males and Females

mf_groups = salary_df.groupby(['Sex'])
men = mf_groups.get_group('Male').count()
women = mf_groups.get_group('Female').count()
men_percent = (men['Name'] / (men['Name'] + women['Name'])) * 100
women_percent = (women['Name'] / (men['Name'] + women['Name'])) * 100
print (" %2.2f" % (men_percent) + " %")
print (" %2.2f" % (women_percent) + " %")
```

```python
# And finally making the plot

label_string = 'Source: BBC, extracted from \'Annex to the BBC Annual Report and Accounts 2016/17\' published 20/07/2017'
label_string = label_string + '\nAvailable at http://downloads.bbc.co.uk/aboutthebbc/insidethebbc/reports/pdf/annex_annual_report_201617.pdf'

colors = {'Male': 'orange', 'Female': 'purple'}
salary_df.plot.bar(x='Name', y=['Band Upper Bound'], figsize=(24, 6), color=[colors[i] for i in salary_df['Sex']])
plt.ylabel('Salary upper bound (£)', fontsize=20)
plt.xlabel('Name', fontsize=20)
plt.title('BBC Salaries for \'On-air Talent and Contributors\' 2016-17\n\n\n', fontsize=30)
plt.gcf().text(0.125, 1, label_string, fontsize=15)

legendpatch_list = ([mpatches.Patch(color='orange', label=("Males (" + "%2.2f" % (men_percent) + " % of individuals)"))]+
                    [mpatches.Patch(color='purple', label=("Females (" + "%2.2f" % (women_percent) + " % of individuals)"))])   
plt.legend(handles=legendpatch_list, loc=(0.76, 0.8))
plt.savefig('vis_output/BBCSalaries.png')
```
